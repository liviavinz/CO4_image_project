```python
# Create grids of images
def random_crop(image):
    h, w = image.shape[:2]
    s_min = min(h, w, 100)
    roi_w = np.random.randint(s_min, w)
    roi_h = np.random.randint(s_min, h)
    roi_x = np.random.randint(0, w-roi_w)
    roi_y = np.random.randint(0, h-roi_h)
    return image[roi_y:roi_y+roi_h, roi_x:roi_x+roi_w]

np.random.seed(42)
imgs = [random_crop(img1) for _ in range(8)]
titles = ["Crop %d" % (i+1) for i in range(8)]

tools.show_header("Default")
tools.show_image_grid(imgs, titles=titles)

tools.show_header("Suppress info")
tools.show_image_grid(imgs, titles=titles, suppress_info=True)

tools.show_header("No frame")
tools.show_image_grid(imgs, titles=titles, shape=None, show_canvas_frame=False)

tools.show_header("Force square frame")
tools.show_image_grid(imgs, titles=titles, shape=None, box_aspect=1)

tools.show_header("With border")
tools.show_image_grid(imgs, titles=titles, shape=None, box_aspect=1, 
                      suppress_info=True,
                      title_kwargs={"color": "red", "weight": "bold"},
                      background_color=None, 
                      show_frame="forced", 
                      frame_color="red", 
                      frame_width=3)
```
